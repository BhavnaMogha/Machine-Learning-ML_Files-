```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# PieChart

```python
areas=['Marketing','Sales','Development','HR','Customer Support']
budget=[2,5,4,10,1]
```

```python
plt.pie(budget, labels=areas, autopct="%0.1f%%")
plt.show()
```

```python
x=[23,45,12,49,97,32,11,44]
y=[45,34,67,87,55,22,89,98]
plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age Vs Salary Graph')
```

```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[233,456,770,120,222,444,200,300]
offline_price=[400,300,100,333,444,565,565,899]
```

```python
plt.plot(product_id,online_price, color='red',marker="D",markersize=10)
plt.plot(product_id,offline_price, color='blue',marker="D",markersize=10)
plt.xlabel("product")
plt.ylabel("prices")
plt.title("product price comparison")
plt.show()
```
