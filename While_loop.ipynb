```python
#Palindrome input no.
n=int(input('enter n'))
rev=0
x=n
while n>0:
  d=n%10
  rev=rev*10+d
  n=n//10
print(rev)
if x==rev:
  print('palindrome')
else:
  print('not palindrome')
```
